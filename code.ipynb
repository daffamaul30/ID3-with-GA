```
def generatePopulation():
    populasi = []
    chromo1 = []
    jml_kromosom = 20
    jml_bit_kromosom = 30
    
    for individu in range(jml_kromosom):
        populasi.append(generateChromosome(jml_bit_kromosom))
    pops = []
    pop = []
    for i in range(len(populasi)):
      pops = []
      for j in range(len(populasi[i])):
        pops.append(populasi[i][j])
      pop.append(pops)

    
    return pop
```

```
def readFile():
    train = open("train.csv", "r")
    dataTrain = train.readlines()
    
    array = []

    # print (len(dataTrain))

    for i in range(1, len(dataTrain)):
      array.append(dataTrain[i].replace('\n','').split(';'))
    # for i in range(0, len(dataTrain)):
    #   print(dataTrain[i])
    return array
```

```
def parent_select(ppls,train,fit):
   
    haha = -1
    for indekss in range(10):
        indekss = random.randint(0,len(ppls)-1)
        # print("INDEKS",indekss)
        fit1 = fit[indekss]
        fit2 = fit[haha]
        if (haha == -1 or (fit1 > fit2)):
            haha = indekss

    return ppls[haha].copy(),ppls[indekss].copy()
pop = generatePopulation()
train = convertToBit()
fit = fitness(pop,train)
ini1,ini2 = parent_select(pop,train,fit)
print(ini1,ini2)
```

```
def crossover(prt1,prt2):
  pelCros = 0.8
  rand = random.uniform(0,1)
  # print("random",rand)
  if rand <= pelCros:
    #p1{point1,point2}
    # print(len(prt1))
    # print(len(prt2))
    point1 = random.randint(1,len(prt1)-13)
    point2 = random.randint(point1+1,(len(prt1))-1)
    # print("HAHAHA")
    arrpo = []
    jrk = 0
    gap = 0
    jrk = point2-point1
    gap = jrk % 15
    # print("POINT1",point1)
    # print("POINT2",point2)
    # print("JRK",jrk)
    # print("GAP",gap)

    # p2{point1,point3}
    # p2{point1,point4}
    # p2{point5,point2}
    # p2{point6,point2}
    point3 = point1 + jrk
    point4 = point1 + gap
    point5 = point2 - jrk
    point6 = point2 - gap
    
    arrpo.append(point3)
    arrpo.append(point4)
    arrpo.append(point5)
    arrpo.append(point6)
    # print(arrpo)
    tmp = prt1[point1:point2].copy()
    tmp2 = prt1[point2:len(prt1)].copy()

    rand = random.randint(0,3)
    # print("RAND",rand)
    if (rand == 0) or (rand == 1):
      tmp3 = prt2[point1:arrpo[rand]].copy()
      tmp4 = prt2[arrpo[rand]:len(prt2)].copy()
    else:
      tmp3 = prt2[arrpo[rand]:point2].copy()
      tmp4 = prt2[point2:len(prt2)].copy()
    if (rand == 0) or (rand == 1):
      # print(point1,',',point2)
      # print(point1,',',arrpo[rand])
      prt1[point1:] = tmp3
      prt1[arrpo[rand]:] = tmp2
      prt2[point1:] = tmp
      prt2[point2:] = tmp4
    else:
      # print(point1,',',point2)
      # print(arrpo[rand],',',point2)
      prt1[point1:] = tmp3
      prt1[point2:] = tmp2
      prt2[arrpo[rand]:] = tmp
      prt2[point2:] = tmp4
  
  pjg1 = len(prt1)
  pjg2 = len(prt2)

  if (pjg1 > 75):
    #print (panj)
    prt1[75:] = []
  if (pjg2 > 75):
    #print (panj)
    prt2[75:] = []
  
  return prt1,prt2


pop = generatePopulation()
train = convertToBit()
fit = fitness(pop,train)
ini1,ini2 = parent_select(pop,train,fit)
print(ini1)
print(ini2)
cr1,cr2 = crossover(ini1,ini2)
print(cr1)
print(cr2)
```

```
def mutation(co1, co2):
    pelMut = 0.3
    rand = random.uniform(0,1)
    # print(rand)
    if rand <= pelMut:
        for i in range(len(co1)):
            indeks = random.randint(0,len(co1))
            if indeks <= len(co1)-1:
                #print("indeks",indeks)
                co1[indeks] = random.randint(0,1)
        for j in range(len(co2)):
            index = random.randint(0,len(co2))
            if index <= len(co2)-1:
                co2[index] = random.randint(0,1)
    
    return co1,co2
pop = generatePopulation()
train = convertToBit()
fit = fitness(pop,train)
ini1,ini2 = parent_select(pop,train,fit)
print(ini1)
print(ini2)
cr1,cr2 = crossover(ini1,ini2)
print(cr1)
print(cr2)
mu1,mu2 = mutation(cr1,cr2)
print(mu1)
print(mu2)
```
